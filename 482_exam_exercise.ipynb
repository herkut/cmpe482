```octave
#classical gram-schmidt
function [Q,R] = gs(A)
    [m,n] = size(A)

    Q = zeros(m,n)
    R = zeros(n,n)

    for j = 1:n
        v = A(:,j)
        for i = 1:j-1
            R(i,j)=Q(:,i)'*A(:,i)
            v=v-R(i,j)*Q(:,i)
        endfor
        R(j,j)=norm((v),2)
        Q(:,j)=v/R(j,j)
    endfor
end
################################
#                              #
################################

#modified gram-schmidt
function [Q,R] = mgs(A)
    [m,n] = size(A)

    Q = zeros(m,n)
    R = zeros(n,n)
    v = zeros(m,n)

    for i = 1:n
        v(:,i) = A(:,i)
    endfor 
    for i = 1:n
        R(i,i) = norm(v(:,i),2)
        Q(:,i) = v(:,i)/R(i,i)
        for j = i+1:n
            R(i,j)=Q(:,i)'*v(:,j)
            v(:,j)=v(:,j)-R(i,j)*Q(:,i)
        endfor
    endfor
end

gs(A)

mgs(A)
```
